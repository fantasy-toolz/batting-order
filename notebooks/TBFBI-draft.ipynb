```python
import pandas as pd
import numpy as np

# Load the CSV file to examine its contents
file_path = "data/2025-TGFBI-Pick-Tracker.csv"
df = pd.read_csv(file_path)

# Display basic information and the first few rows of the dataset
df.info(), df.head()
```

```python
df_clean = df  # Remove header row duplication
df_clean["ADP"] = pd.to_numeric(df_clean["ADP"], errors="coerce")
df_clean["Min"] = pd.to_numeric(df_clean["Min"], errors="coerce")
df_clean["Max"] = pd.to_numeric(df_clean["Max"], errors="coerce")
df_clean["Range"] = pd.to_numeric(df_clean["Range"], errors="coerce")
df_clean = df_clean.loc[df_clean['Min']>0]  # Remove header row duplication

# Sort by ADP to see top-ranked players
df_clean_sorted = df_clean.sort_values(by="Min")
```

```python
#996-1021

for lg in range(996,1021):
    df_clean[str(lg)] = pd.to_numeric(df_clean[str(lg)], errors="coerce")
```

```python
df_clean.index
```

```python
df_clean['MDP'] = np.zeros(df_clean['ADP'].values.size)

for indx,mdp in enumerate(df_clean['MDP'].values):
    df_clean['MDP'].values[indx] = np.nanmedian(np.array([df_clean[str(lg)].values[indx] for lg in range(996,1021)]))

```

```python
df_clean['DPdiff'] = df_clean['ADP']-df_clean['MDP']
```

```python
# negative means guys reached (adp is higher than median)

df_clean_sorted = df_clean.sort_values(by="DPdiff")
df_clean_sorted.head(10)
```

```python
# positive means the tail is really long (i.e. guys waited)

df_clean_sorted = df_clean.sort_values(by="DPdiff",ascending=False)
df_clean_sorted.head(10)
```
